# Problem Statement

The objective of this task is to detect hate speech in tweets. For the sake of simplicity, we say a tweet contains hate speech if it has a racist or sexist sentiment associated with it. So, the task is to classify racist or sexist tweets from other tweets.


Formally, given a training sample of tweets and labels, where label '1' denotes the tweet is racist/sexist and label '0' denotes the tweet is not racist/sexist, your objective is to predict the labels on the test dataset.

**Evaluation Metric - $F_{1}$ Score** 

$F_1 = \frac{2 * Precision * Recall}{Precision + Recall}$

# Tweets Preprocessing

## Removing Twitter Handles

Since all the handles are masked (since represented as @user), they're not conveying any important information anyways. Hence remove them straightaway from our document corpus.

```python
pattern = r'@[A-Za-z]+'
tweets_wo_user = []

for i in train.tweet:
    tweets_wo_user.append(re.sub(pattern, "", i))

train['clean_tweet'] = tweets_wo_user

tweets_wo_user = []
for j in test.tweet:
    tweets_wo_user.append(re.sub(pattern, "", j))

test['clean_tweet'] = tweets_wo_user
```

```python
# Replace contractions from train tweets
expanded_tweet = []
for i in train.clean_tweet:
    container = []
    for token in i.split():
        if token in CONTRACTION_MAP.keys():
            container.append(CONTRACTION_MAP[token])
        else:
            container.append(token)
    expanded_tweet.append(" ".join(container))
train.clean_tweet = expanded_tweet
```

```python
# Replace contractions from test tweets
expanded_tweet = []
for i in test.clean_tweet:
    container = []
    for token in i.split():
        if token in CONTRACTION_MAP.keys():
            container.append(CONTRACTION_MAP[token])
        else:
            container.append(token)
    expanded_tweet.append(" ".join(container))
test.clean_tweet = expanded_tweet
```

## Removing Punctuations and Special Characters

All these three can prove to be good indicators of sentiment. 

Or whenever someone's angry and they want to emote things, they add punctuations like exclamation or questions for being rhetoric and so on.

But in our simplistic analysis, we won't need all those and so, let's get rid of them. 

```python
pattern = r'[\(\)\[\]\.\,\:\;\'\"]'
train.clean_tweet = train.clean_tweet.str.replace(pattern, "")
test.clean_tweet = test.clean_tweet.str.replace(pattern, "")
```

```python
train.head(2)
```

```python
test.head(2)
```

## Replace all the digits by the word digit

When someone's reviewing a product and it's expensive, s/he might add the price of the product and claim it's expensive and cuss a lot in the review. It could be the other way round as well. But keeping all the digits would increase the feature-space by a large amount.

Instead we can replace any digit or string of digits with the word digit and then analyze it, it would help keep the meaning intact to some extent.

```python
train.clean_tweet = train.clean_tweet.map(lambda x: re.sub(r"[0-9]+", "digit", str(x)))
test.clean_tweet = test.clean_tweet.map(lambda x: re.sub(r"[0-9]+", "digit", str(x)))
```

## Extraction of all the hashtags

Hashtags are crucial features which can help tell us categorize tweets nicely. Hence extract them from the raw tweets.

It could be helpful in exploratory data analysis of the tweets.

```python
pattern = r'#([A-za-z0-9]+)'
hashtags = []
for i in train.tweet:
    hashtags.append(re.findall(pattern, i))
train['hashtags'] = hashtags

hashtags = []
for j in test.tweet:
    hashtags.append(re.findall(pattern, j))
test['hashtags'] = hashtags
```

## Word Cleaning

### Remove short length words

Words which are very short like hmm, oh, the, is, are etc. don't add much meaning to our words. Let's therefore remove such short words whose length is less than 3 words.

### Remove stop words

Words like the, and, is, an, within etc. occur very frequently and don't add much meaning to the sentences in a semantic way. Hence do not keep such words in the corpus.

### Remove non-alphabet text

Also, let's clean the other emoticons or utf-8 non-alphabetic words from the text since they're not of any use to us.

### Express words in their base form

Finally let's lemmatize the words to their base form eg. (is, was, been) = be; (knowing, knower, known) = know and so on ans simultaneously normalize the words i.e. make every word lowercase.

```python
train_tweets = []
for doc in nlp.pipe(train.clean_tweet):
    container = []
    for token in doc:
        # Check if text is word or not and remove emoticons, url links, email links and other utf-8 characters
        # Do this by keeping only ascii characters since emoticons and other stuff are only in utf-8
        if not(token.is_punct) and not(token.is_stop) and not(token.like_url) and not(token.like_email) and(token.is_ascii):
            # Lemmatize the word to it's base form and then append it
            if token.lemma_ != "-PRON-":
                container.append(token.lemma_.lower())
            else:
                container.append(token.text.lower())
    train_tweets.append(" ".join(container))
train.clean_tweet = train_tweets
train.head(3)
```

```python
test_tweets = []
for doc in nlp.pipe(test.clean_tweet):
    container = []
    for token in doc:
        if not(token.is_punct) and not(token.is_stop) and not(token.like_url) and not(token.like_email) and (token.is_ascii):
            if token.lemma_ != "-PRON-":
                container.append(token.lemma_.lower())
            else:
                container.append(token.text.lower())
    test_tweets.append(" ".join(container))
test.clean_tweet = test_tweets
test.head(3)
```

## Adding Number of Hashtags, tweet_length, meaning_fraction as a feature

It is my observation that people who tend to be highly polar about their opinion tend to use a lot of hashtags since they want to assert their opinion very strongly and consequently their tweets tend to be larger. Let's try to use this observation and see if it helps.

Meaning Fraction is an experimental feature which I am introducing. It is the ratio of length of clean tweet to that of the raw tweet. Let's see if this feature helps in prediction.

```python
# Number of hashtags
train['No_hashtags'] =  train.hashtags.apply(lambda x: len(x))
test['No_hashtags'] = test.hashtags.apply(lambda x: len(x))

# Apply len to raw tweet and not clean tweet since we're interested in actual tweet length
train['Tweet_len'] = train.tweet.apply(lambda x: len(x))
test['Tweet_len'] = test.tweet.apply(lambda x: len(x))
```

```python
train_meaning = []
for i in train.index:
    train_meaning.append(len(train.clean_tweet[i]) / (1 + len(train.tweet[i]))) # Add 1 to dr to avoid div by 0 error
    
test_meaning = []
for i in test.index:
    test_meaning.append(len(test.clean_tweet[i]) / (1 + len(test.tweet[i])))

train['Meaning_frac'] = train_meaning
test['Meaning_frac'] = test_meaning
```

# Hashtag analysis of labelled Tweets

Let's analyze the hashtags of labelled tweets to find out which hashtag was most associated with negative tweets and which one was most associated with positive tweets.

```python
pos_hashtags = {}
neg_hashtags = {}

positive_hashtags = train[train.label == 0].hashtags
negative_hashtags = train[train.label == 1].hashtags

for i in positive_hashtags:
    for hashtag in i:
        word = hashtag.lower()
        if word in pos_hashtags.keys():
            pos_hashtags[word] = pos_hashtags[word] + 1
        else:
            pos_hashtags[word] = 1

for i in negative_hashtags:
    for hashtag in i:
        word = hashtag.lower()
        if word in neg_hashtags.keys():
            neg_hashtags[word] = neg_hashtags[word] + 1
        else:
            neg_hashtags[word] = 1            
```

```python
top_pos_hashtags = sorted(pos_hashtags, key = lambda x: pos_hashtags[x], reverse = True)[:15]
top_neg_hashtags = sorted(neg_hashtags, key = lambda x: neg_hashtags[x], reverse = True)[:15]

freq_top_pos_hashtags = [pos_hashtags[i] for i in top_pos_hashtags]
freq_top_neg_hashtags = [neg_hashtags[i] for i in top_neg_hashtags]
```

```python
fig, axes = plt.subplots(1,2, figsize = (20,7))

sns.barplot(x = top_pos_hashtags, y = freq_top_pos_hashtags, ax = axes[0])
axes[0].title.set_text("Positive Tweet Labels")
axes[0].tick_params(axis = "x", rotation = 90)

sns.barplot(x = top_neg_hashtags, y = freq_top_neg_hashtags, ax = axes[1])
axes[1].title.set_text("Negative Tweet Labels")
axes[1].tick_params(axis = "x", rotation = 90)
```

```python
def add_feature(X, feature_to_add):
    
    # Returns sparse feature matrix with added feature.
    # Feature_to_add can also be a list of features.
    
    from scipy.sparse import csr_matrix, hstack
    return hstack([X, csr_matrix(feature_to_add).T], 'csr')
```

# Bag Of Words

We need to create a bag of words or an inverted index from these given documents. In order to accomplish this task, we will fit a TfidfVectorizer on our train data and use it to transform our validation and test data accordingly. 

Also we will keep the terms which appear in a minimum of five documents in order to avoid extreme sparsity in our train and test matrices.

```python
from sklearn.feature_extraction.text import TfidfVectorizer, TfidfTransformer

# min_df - Minimum no of Documents the word must occur in to be considered as a term
# smooth_idf - It adds 1 to doc frequencies and prevents division by 0
# sublinear_tf - Replace tf with 1 + log(tf) [Note min_df = 3 so, will never encounter log(0)]
# strip_accents - Remove those naive', le' miserable' etc. kind of accents from the document
# ngram - add 
# Token pattern - It should start with a word and have at least one character

TFVect = TfidfVectorizer(min_df = 3, ngram_range=(1,5), smooth_idf=True, stop_words="english", 
                         max_features=None, sublinear_tf = True, strip_accents='unicode', 
                         analyzer = 'word', token_pattern=r'\w{1,}')

TFVect.fit(X_train)

X_tr = TFVect.transform(X_train)
X_vl = TFVect.transform(X_val)
X_test = TFVect.transform(test.clean_tweet)

# For normalization sake one can use this.
# It's optional not compulsory
tfidf_transformer = TfidfTransformer()
X_tr = tfidf_transformer.fit_transform(X_tr)
X_vl = tfidf_transformer.transform(X_vl)
X_test = tfidf_transformer.transform(X_test)
```

```python
# Adding tweet length and number of hashtags to train, validation and test data
# After trial and error, it was observed that only meaning frac considerably affects 
# The f-1 score and not any other parameter. Hence, add only that and not other params

# Tweet_lengths = list(train_features['Tweet_len'])
# Hashtag_Num = list(train_features['No_hashtags'])
Meaning_frac = list(train_features['Meaning_frac'])
#X_tr = add_feature(X_tr, Tweet_lengths)
#X_tr = add_feature(X_tr, Hashtag_Num)
X_tr = add_feature(X_tr, Meaning_frac)

# Tweet_lengths = list(val_features['Tweet_len'])
# Hashtag_Num = list(val_features['No_hashtags'])
Meaning_frac = list(val_features['Meaning_frac'])
#X_vl = add_feature(X_vl, Tweet_lengths)
#X_vl = add_feature(X_vl, Hashtag_Num)
X_vl = add_feature(X_vl, Meaning_frac)


# Tweet_lengths = list(test_features['Tweet_len'])
# Hashtag_Num = list(test_features['No_hashtags'])
Meaning_frac = list(test_features['Meaning_frac'])
# X_test = add_feature(X_test, Tweet_lengths)
# X_test = add_feature(X_test, Hashtag_Num)
X_test = add_feature(X_test, Meaning_frac)
```

# Model Building and Evaluation

## Simple Naive Bayes Classifier

```python
from sklearn.naive_bayes import MultinomialNB

NB_clfr = MultinomialNB().fit(X_tr, y_train)

y_pred_NB = NB_clfr.predict(X_vl)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, f1_score, recall_score, classification_report, precision_score

f1_NB = f1_score(y_val, y_pred_NB)
recall_NB = recall_score(y_val, y_pred_NB)
precision_NB = precision_score(y_val, y_pred_NB)

NB_metrics = {'f1-Score':f1_NB, 'Recall':recall_NB, 'Precision':precision_NB}
#print(np.round(NB_metrics, 2))
print("\nPerformance of Naive Bayes Classifier is as follows:\n\n{}".format(classification_report(y_val, y_pred_NB)))
```

## Logistic Regression Classifier

### Simple Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LR_clfr = LogisticRegression(C = 10, solver = 'liblinear').fit(X_tr, y_train)

# Predict class probabilities
y_pred_proba = LR_clfr.predict_proba(X_vl)

# Study the distribution of probabilities for positive class
print(pd.Series(y_pred_proba[:,1]).describe())
rcParams['figure.figsize'] = 10,5
sns.distplot(y_pred_proba[:,1], rug = True, hist = True, kde = True)
plt.title("Probability Distribution of Validation data")
plt.show()

# Plot the receiver operating characteristics to decide a threshold value
fpr, tpr, threshold = roc_curve(y_val, y_pred_proba[:,1], pos_label = 1)
print("AUC Score {:.3f}\n".format(roc_auc_score(y_val, y_pred_proba[:,1])))
rcParams['figure.figsize'] = 15,5
lineplot = sns.lineplot(x = fpr, y = tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristics Curve")
for i in range(0, len(threshold)):
    if i % 20 == 0:
        lineplot.text(fpr[i], tpr[i] + 0.03 * (-1)**(i/20), round(threshold[i], 2), rotation = 90)
plt.show()


# Make predictions based on a threshold
Threshold = 0.52
y_val_pred_LR = [1 if i > Threshold else 0 for i in y_pred_proba[:,1]]

# Evaluate the predictions
f1_LR = f1_score(y_val, y_val_pred_LR)
recall_LR = recall_score(y_val, y_val_pred_LR)
precision_LR = precision_score(y_val, y_val_pred_LR)

LR_metrics = {'f1-Score':f1_LR, 'Recall':recall_LR, 'Precision':precision_LR}
#print(np.round(LR_metrics, 2))
print("\nPerformance of LR Classifier is as follows:\n\n{}".format(classification_report(y_val, y_val_pred_LR)))
```

### Logistic Regression With SMOTING
